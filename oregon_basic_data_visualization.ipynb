```python
# geojson mapping
import geopandas as gpd
import matplotlib.pyplot as plt

# Load  GeoJSON file
file_path = r"C:\Users\addis\OneDrive\COLLEGE\!Fall2025\Geog573\final_project\oregon_fire_bounds.geojson"
gdf = gpd.read_file(file_path)

# Quick check
print(f'data head: {gdf.head()}')

# Basic plot
gdf.plot(edgecolor='black', facecolor='orange', figsize=(10, 10))
plt.title("Oregon Fire Boundaries")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
